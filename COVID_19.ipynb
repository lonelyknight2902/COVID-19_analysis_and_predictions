```python
#For tabular dataset
import pandas as pd
#Visualization libraries
pd.plotting.register_matplotlib_converters()
import matplotlib.pyplot as plt
#Tell Python to visualize in this window, not creating new one
%matplotlib inline
import seaborn as sns
```

#**Task 1: Data Collection**



```python
#Reading data
df_asia = pd.read_csv("/content/Asia.csv", parse_dates=['date'])
df_americas = pd.read_csv("/content/Americas.csv", parse_dates=['date'])
df_africa = pd.read_csv("/content/Africa.csv", parse_dates=['date'])
df_europe = pd.read_csv("/content/Europe.csv", parse_dates=['date'])
```

#**Task 2: Data Analysis**

#**1.Asia**

```python
countries = df_asia.country.unique()
df_asia.set_index(['country', 'date'], inplace=True)
plt.figure(figsize=(20,10))
sns.set_style("whitegrid")
for idx in countries:
  if idx == "India":
    continue
  sns.lineplot(data=df_asia.xs(idx)['cumulative_total_cases'], label = idx, markers=True)
plt.title("Cumulative total cases in 7 Asian countries")
plt.xlabel("Date")
plt.ylabel("Cumulative total cases")
plt.xlim("2020-07-01","2021-05-31")
plt.ylim(0, 800000)
```

```python
#Since India figures is very high so we decided to draw it in seperate graph
plt.figure(figsize=(20,10))
sns.set_style("whitegrid")
sns.lineplot(data=df_asia.xs("India")['cumulative_total_cases'], markers=True)
plt.title("Cumulative total cases in India")
plt.xlabel("Date")
plt.ylabel("Cumulative total cases")
plt.xlim("2020-07-01","2021-05-31")
plt.ylim(0, 3e7)
plt.yticks([0,0.5e7,1.0e7,1.5e7,2.0e7,2.5e7, 3e7],['0','5M','10M','15M','20M','25M','30M'])
```

#**2.Europe**

```python
countries = df_europe.country.unique()
df_europe.set_index(['country', 'date'], inplace=True)
plt.figure(figsize=(20,10))
sns.set_style("whitegrid")
for idx in countries:
  sns.lineplot(data=df_europe.xs(idx)['cumulative_total_cases'], label = idx, markers=True)
plt.title("Cumulative total cases in 9 European countries")
plt.xlabel("Date")
plt.ylabel("Cumulative total cases")
plt.xlim("2020-07-01","2021-05-31")
plt.ylim(0, 6e6)
plt.yticks([0,1e6,2e6,3e6,4e6,5e6,6e6],['0','1M','2M','3M','4M','5M','6M'])
```
