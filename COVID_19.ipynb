##**3.1 AutoRegression**

Autoregression is a time series model that uses observations from previous time steps as input to a regression equation to predict the value at the next time step.

```python
from statsmodels.tsa.ar_model import AutoReg
from math import sqrt
from sklearn.metrics import mean_squared_error
```

###**3.1.a Global prediction**

```python
global_pdf = pd.read_csv('/content/worldometer_coronavirus_daily_data.csv', parse_dates=['date'])
cumulative_total_cases = global_pdf.groupby('date').sum()['cumulative_total_cases'].reset_index()
cumulative_total_cases.columns = ['date','cumulative_total_cases']
cumulative_total_cases['date'] = pd.to_datetime(cumulative_total_cases['date'])
```

```python
series = cumulative_total_cases.set_index('date')['cumulative_total_cases']
X = series.loc['2020-07-01':'2021-05-31'].values
# train autoregression
model = AutoReg(X, lags=29)
model_fit = model.fit()
# make predictions
predictions = model_fit.predict(start=len(X), end=len(X)+92-1, dynamic=False)
df = pd.DataFrame({ 'date': pd.date_range('2021-06-01', periods=92, freq='D'), 'Predicted' : predictions})
df = df.set_index('date')
plt.figure(figsize=(20,10))
sns.set_style("whitegrid")
sns.lineplot(data=series.loc['2020-07-01':'2021-05-31'], label='Past')
sns.lineplot(data=df, palette=['red'])
title = "Global cumulative total cases and prediction from July 2020 to August 2021"
plt.title(title)
plt.xlabel("Date")
plt.ylabel("Cumulative total cases")
plt.xlim('2020-07-01','2021-09-01')
plt.ylim(0, 2.5e8)
plt.yticks([0,0.5e8,1e8,1.5e8,2e8,2.5e8], ['0','50M','100M','150M','200M','250M'])
```

##**3.2 Prophet**

Prophet is a procedure for forecasting time series data based on **an additive model** where non-linear trends are fit with yearly, weekly, and daily seasonality, plus holiday effects. It works best with time series that have strong seasonal effects and several seasons of historical data. Prophet is robust to missing data and shifts in the trend, and typically handles outliers well.

```python
from fbprophet import Prophet
```

**3.2.a Global prediction**

```python
global_pdf = pd.read_csv('/content/worldometer_coronavirus_daily_data.csv', parse_dates=['date'])
cumulative_total_cases = global_pdf.groupby('date').sum()['cumulative_total_cases'].reset_index()
cumulative_total_cases.columns = ['ds','y']
cumulative_total_cases['ds'] = pd.to_datetime(cumulative_total_cases['ds'])
cumulative_total_cases
```

```python
m = Prophet(interval_width=0.95)
m.fit(cumulative_total_cases)
future = m.make_future_dataframe(periods=365)
forecast = m.predict(future)
confirmed_forecast_plot = m.plot(forecast)
```
