```python
import pandas as pd
pd.plotting.register_matplotlib_converters()
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
data_path = "/content/Europe.csv"
data = pd.read_csv(data_path, parse_dates=['date'])
countries = data.country.unique()
```

```python
data.set_index(['country', 'date'], inplace=True)
data
```

```python
plt.figure(figsize=(20,10))
sns.set_style("whitegrid")
for idx in countries:
  sns.lineplot(data=data.xs(idx)['cumulative_total_cases'], label = idx, markers=True)
plt.title("Cumulative total cases in 9 European countries")
plt.xlabel("Date")
plt.ylabel("Cumulative total cases")
```
