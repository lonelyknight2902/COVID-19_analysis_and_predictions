# Task 3: Model prediction

##**3.2 AutoRegression**

```python
from statsmodels.tsa.ar_model import AutoReg
from math import sqrt
from pandas.plotting import lag_plot
from sklearn.metrics import mean_squared_error
```

```python
countries = df_europe.reset_index().country.unique()
for country in countries:
  series = df_europe.xs(country).reset_index().set_index('date')
  series = series['cumulative_total_cases']
  X = series.values
  # train autoregression
  model = AutoReg(X, lags=29)
  model_fit = model.fit()
  #print('Coefficients: %s' % model_fit.params)
  # make predictions
  predictions = model_fit.predict(start=len(X), end=len(X)+92-1, dynamic=False)
  # plot results
  #pyplot.plot(test)
  #pyplot.plot(predictions, color='red')
  #pyplot.show()
  df = pd.DataFrame({ 'date': pd.date_range('2021-06-01', periods=92, freq='D'), 'Predicted' : predictions})
  df = df.set_index('date')
  plt.figure(figsize=(20,10))
  sns.set_style("whitegrid")
  sns.lineplot(data=series, label='Past')
  sns.lineplot(data=df, palette=['red'])
  title = "Cumulative total cases and prediction in " + country + " from January 2021 to September 2021"
  plt.title(title)
  plt.xlabel("Date")
  plt.ylabel("Cumulative total cases")
  plt.xlim('2021-01-01','2021-08-31')
  plt.ylim(series.min(), predictions.max() + 1e6)
```

##**3.1 Prophet**

Prophet is a procedure for forecasting time series data based on an additive model where non-linear trends are fit with yearly, weekly, and daily seasonality, plus holiday effects. It works best with time series that have strong seasonal effects and several seasons of historical data. Prophet is robust to missing data and shifts in the trend, and typically handles outliers well.

```python
from fbprophet import Prophet
```

```python

#predict data frame = pdf
pdf = pd.read_csv("/content/Asia_prophet.csv", parse_dates=['date'])
pdf = pdf.rename(columns={'date': 'ds', 'cumulative_total_cases':'y'})
#rename for prophet
asia_countries_grouped = pdf.groupby('country')
for g in asia_countries_grouped.groups:
    group = asia_countries_grouped.get_group(g)
    m = Prophet()
    m.fit(group)
    future = m.make_future_dataframe(periods=90)
    forecast = m.predict(future)
    print(g)
    confirmed_forecast_plot = m.plot(forecast)
    #print(confirmed_forecast_plot)
```
